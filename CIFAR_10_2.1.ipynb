```python
#importing the required libraries
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
```

```python
#Loading the dataset into training and testing dataset
training_data = datasets.CIFAR10(
    root = "data",
    train = True,
    download = True,
    transform = ToTensor()
)
```

```python
testing_data = datasets.CIFAR10(
    root = "data",
    train = True,
    download = True,
    transform = ToTensor()
)
```

```python
#preparing the dataloaders
train_data = DataLoader(training_data, batch_size = 64, shuffle = True)
test_data = DataLoader(testing_data, batch_size = 64, shuffle = True)
```

```python
class NeuralNetwork(nn.Module):
  def __init__(self):
    super(NeuralNetwork,self).__init__()
    self.flatten = nn.Flatten()
    self.layer1 = nn.Linear(32*32*3,512)
    self.layer2 = nn.Linear(512,100)
    self.layer3 = nn.Linear(100,10)
    self.activ = nn.ReLU()
    self.norma = nn.BatchNorm1d(512)
    self.norma2 = nn.BatchNorm1d(100)
    self.dropout = nn.Dropout(0.2)
  def forward(self,x):
    x = self.flatten(x)
    x = self.layer1(x)
    x = self.norma(x)
    x = self.activ(x)
    x = self.layer2(x)
    x = self.norma2(x)
    x = self.activ(x)
    x = self.layer3(x)
    return x
```

```python
model = NeuralNetwork()
learning_rate = 1e-3
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    for batch, (X, y) in enumerate(dataloader):
        # Compute prediction and loss
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")


def test_loop(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate,weight_decay= 1e-4,momentum=0.9)

epochs = 15
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(train_data, model, loss_fn, optimizer)
    test_loop(test_data, model, loss_fn)
print("Done!")  
```
